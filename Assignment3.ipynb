### 1. Provide summary statistics (mean, median, minimum, maximum, standard deviation) for a dataset (age, income etc.) with numeric variables grouped by one of the qualitative (categorical) variable. For example, if your categorical variable is age groups and quantitative variable is income, then provide summary statistics of income grouped by the age groups. Create a list that contains a numeric value for each response to the categorical variable.

```python
# mean, median, mode, standard deviation, minimum maximum of math score by grouping gender

print(data.groupby('gender')['math score'].describe())
print("\n\nMedian\n")
print(data.groupby('gender')['math score'].median())
print("\n\nMode\n")
print(data.groupby('gender')['math score'].agg(pd.Series.mode))
```

```python
# mean, median, mode, standard deviation, minimum maximum of reading score by grouping gender

print(data.groupby('gender')['reading score'].describe())
print("\n\nMedian\n")
print(data.groupby('gender')['reading score'].median())
print("\n\nMode\n")
print(data.groupby('gender')['reading score'].agg(pd.Series.mode))
```

```python
# mean, median, mode, standard deviation, minimum maximum of writing score by grouping gender

print(data.groupby('gender')['writing score'].describe())
print("\n\nMedian\n")
print(data.groupby('gender')['writing score'].median())
print("\n\nMode\n")
print(data.groupby('gender')['writing score'].agg(pd.Series.mode))
```

```python
# mean, median, standard deviation, minimum maximum of math score by grouping race/ethnicity

print(data.groupby('race/ethnicity')['math score'].describe())
print("\n\nMedian\n")
print(data.groupby('race/ethnicity')['math score'].median())
```

```python
# mean, median, standard deviation, minimum maximum of reading score by grouping race/ethnicity

print(data.groupby('race/ethnicity')['reading score'].describe())
print("\n\nMedian\n")
print(data.groupby('race/ethnicity')['reading score'].median())
```

### 2. Write a Python program to display some basic statistical details like percentile, mean, standard deviation etc. of the species of ‘Iris-setosa’, ‘Iris-versicolor’ and ‘Iris-virginica’ of iris.csv dataset.

```python
# Load the Iris dataset from iris.csv file
iris = pd.read_csv('iris.csv')
iris.head(10)
```

```python
# basic statistical details like percentile, mean, standard deviation of the species of ‘Iris-setosa’,
# ‘Iris-versicolor’ and ‘Iris-virginica’ using describe method

setosa = iris[iris['Species'] == 'Iris-setosa'].describe()
versicolor = iris[iris['Species'] == 'Iris-versicolor'].describe()
virginica = iris[iris['Species'] == 'Iris-virginica'].describe()


print("\nIRIS-SETOSA")
print(setosa)
print("\nIRIS-VERSICOLOR")
print(versicolor)
print("\nIRIS-VIRGINICA")
print(virginica)
```

```python
setosa = iris[iris['Species'] == 'Iris-setosa'].median()
versicolor = iris[iris['Species'] == 'Iris-versicolor'].median()
virginica = iris[iris['Species'] == 'Iris-virginica'].median()

print("\nMEDIAN\n")

print("\nIRIS-SETOSA")
print(setosa)
print("\nIRIS-VERSICOLOR")
print(versicolor)
print("\nIRIS-VIRGINICA")
print(virginica)
```

```python
setosa = iris[iris['Species'] == 'Iris-setosa']
versicolor = iris[iris['Species'] == 'Iris-versicolor'].agg(pd.Series.mode)
virginica = iris[iris['Species'] == 'Iris-virginica'].agg(pd.Series.mode)

setosa = setosa['PetalLengthCm'].agg(pd.Series.mode)
versicolor = versicolor['PetalLengthCm'].agg(pd.Series.mode)
virginica = virginica['PetalLengthCm'].agg(pd.Series.mode)

print("\nMODE\n")

print("\nIRIS-SETOSA")
print(setosa)
print("\nIRIS-VERSICOLOR")
print(versicolor)
print("\nIRIS-VIRGINICA")
print(virginica)
```
