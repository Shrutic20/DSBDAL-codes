```python
# importing the header files

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# loading the dataset and printing first 5 rows of the 'titanic' dataset

data = pd.read_csv("train.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
# replacing the NA values with the mean values for Age attribute

mean_ = data['Age'].mean()
data['Age'].fillna(value = mean_, inplace = True)
```

```python
data.isnull().sum()
```

```python
sns.boxplot(x = 'Sex', y = 'Age', hue = 'Survived', data = data)
```

```python
sns.stripplot(data = data, y = "Age", x = "Survived")
```

```python
sns.jointplot(data = data, y = "Survived", x = "Age")
```
